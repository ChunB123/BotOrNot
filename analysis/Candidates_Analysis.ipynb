# Best Bot Candidates Selection Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = {
    "Model": [
        "Phi-tweet-100k",
        "Phi-tweet-100k 3-shot",
        "Phi-tweet-50k",
        "Phi-tweet-50k 3-shot",
        "Phi-3.5-mini 3.8B",
        "Phi-3-small 7.4B",
        "Phi-3-medium 14B",
        "Phi-3.5-MoE 42B"
    ],
    "Success fooling rate (%)": [25, 41, 28, 30, 4, 25, 32, 19]
}
```

```python
# Convert to DataFrame
df = pd.DataFrame(data)

# Plotting: Bar Chart
plt.figure(figsize=(12, 6))
colors = ['skyblue' if 'tweet-' in model else 'lightgreen' for model in df["Model"]]
plt.barh(df["Model"], df["Success fooling rate (%)"], color=colors, edgecolor='black')

# Chart details
plt.title("Best Bot Candidate Selection Results", fontsize=14)
plt.xlabel("Success fooling rate (%)", fontsize=12)

plt.xlim(0, 45)  # Adjust axis for better presentation
plt.grid(axis="x", linestyle="--", alpha=0.6)
plt.tight_layout()

# Annotating values on bars
for index, value in enumerate(df["Success fooling rate (%)"]):
    plt.text(value + 1, index, f"{value}%", va='center', fontsize=10)

# Display the chart
plt.show()
```
