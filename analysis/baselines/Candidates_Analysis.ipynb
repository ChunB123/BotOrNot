# Best Bot Candidates Selection Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = {
    "Model": [
        "Phi-tweet-100k (3.8B Zero-shot)",
        "Phi-tweet-100k (3.8B Three-shot)",
        "Phi-tweet-50k (3.8B Zero-shot)",
        "Phi-tweet-50k (3.8B Three-shot)",
        "Phi-3.5 (3.8B)",
        "microsoft/Phi-3-small-8k-instruct (7.4B)",
        "microsoft/Phi-3-medium-128k-instruct (14B)",
        "microsoft/Phi-3.5-MoE-instruct (42B)"
    ],
    "Performance (%)": [25, 41, 28, 30, 4, 25, 32, 19]
}
```

```python
# Convert to DataFrame
df = pd.DataFrame(data)

# Plotting: Bar Chart
plt.figure(figsize=(12, 6))
colors = ['skyblue' if 'tweet-' in model else 'lightgreen' for model in df["Model"]]
plt.barh(df["Model"], df["Performance (%)"], color=colors, edgecolor='black')

# Chart details
plt.title("Best Bot Candidate Selection Results", fontsize=14)
plt.xlabel("Performance (%)", fontsize=12)
plt.ylabel("Model", fontsize=12)
plt.xlim(0, 50)  # Adjust axis for better presentation
plt.grid(axis="x", linestyle="--", alpha=0.6)
plt.tight_layout()

# Annotating values on bars
for index, value in enumerate(df["Performance (%)"]):
    plt.text(value + 1, index, f"{value}%", va='center', fontsize=10)

# Display the chart
plt.show()
```
